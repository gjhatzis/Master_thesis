```python
# Import necessary libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Main reviews' dictionairy and ratings,text and recommendation lists
review_list= {}
review_ratings = []
review_texts = []
review_recommendations = []

# Iterate through all review pages of a specific airline company
for i in range(1,62):
    url = 'https://www.airlinequality.com/airline-reviews/aegean-airlines/page/%s/' %(i)

    r = requests.get(url)

# Check for any status problems
    status = r.status_code
    if status == 200:
        page_num = int(i)
        print('PAGE %s REVIEWS\n'%(page_num))
        
# Gets the html content   
        html_content = r.text 
        html_soup = BeautifulSoup(html_content, "html.parser")
        
# Gets the main content of a review
        review_soup = html_soup.find_all('article', itemprop="review")
        
# Iterates through every review in the page
        for review in range(0,len(review_soup)):
            review_index = review_soup[review]
            
# Finds and gets every review's rating and cleans it
            review_rating = review_index.find('div', class_="rating-10").text
            review_rating_clean= review_rating.replace("\n",'').replace("\r",'')
        
# Finds and gets every review's natural language text and cleans it                   
            review_text = review_index.find('div', class_='text_content').text
            review_clean = review_text.replace('✅ Trip Verified |','').replace('Not Verified |','').replace('✅ Verified Review |','')
            
# Finds and gets every review's recommendation answer           
            review_recommendation_soup = review_index.find('table', class_="review-ratings")
            review_recommendation = review_index.find('td', class_=["review-value rating-no","review-value rating-yes"]).text

# Adds each review's rating, text and recommendation in their lists      
            review_ratings.append(review_rating_clean)
            review_texts.append(review_clean)
            review_recommendations.append(review_recommendation)
            
# Adds each list in the main dictionairy         
        review_list['Airline Name'] = 'Aegean Airlines'
        review_list["Rating"] = review_ratings
        review_list["Text"] = review_texts
        review_list['Recommended'] = review_recommendations


    else:     
        break
```

```python
aegean_df= pd.DataFrame(review_list)
```

```python
# Import NLTK

import nltk
import string
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk import word_tokenize

# Convert to lowercase

new_list_lower = []

for x in df['Text']:
    text_lower = x.lower()
    new_list_lower.append(text_lower)
    
df['Text'] = new_list_lower
```

```python
new_list_punc= []

for x in df['Text']:
    text_p = "".join([char for char in x if char not in string.punctuation])
    new_list_punc.append(text_p)
    
df['Text'] = new_list_punc
```

```python
df["Text"]
```

```python
new_list_token = []

for x in df['Text']:
    words= word_tokenize(x)
    new_list_token.append(words)

df['Text'] = new_list_token
```

```python
df["Text"]
```

```python
stop_words = stopwords.words('english')
new_list_stop= []

for x in df['Text']:
    filtered_words = [word for word in x if word not in stop_words]
    new_list_stop.append(filtered_words)
df['Text'] = new_list_stop
```
