```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
review_list= []
total_reviews = 0
for i in range(1,201):
    url = 'https://www.trustpilot.com/review/www.ryanair.com?page=%s' %(i)

    r = requests.get(url)
    
    csv_file = open('ryan_air','wb')
    status = r.status_code
    if status == 200:
        page_num = int(i)
        print('PAGE %s REVIEWS\n'%(page_num))
       
        html_content = r.text
        html_soup = BeautifulSoup(html_content, "html.parser")
        
        result_soup = html_soup.find_all('p',class_= "review-content__text")
        
        for review in range(1,len(result_soup)):
            total_reviews +=1
            
            #print('Review no.%s\n'%(review))
            review_list.append(result_soup[review].text)
            #print(result_soup[review].text)
            
    else:     
        break
df_ryan = pd.DataFrame(review_list)
print('THE TOTAL REVIEWS ARE: '+str(total_reviews))
```

```python
df_ryan.to_excel('rewiews2.xlsx')
```
