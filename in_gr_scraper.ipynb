```python
import requests
from bs4 import BeautifulSoup

#crawls through the 5 first pages of the site
for num_page in range(1,5):
    url = 'https://www.in.gr/latestnews/page/%s/'%(num_page)
    r = requests.get(url)
    html_soup = BeautifulSoup(r.text,'html.parser')
    articles = []
#gets all articles from each page
    for i in range(1,31):
        for item in html_soup.find_all('li', class_='mm_%s' % (i)):
            item_a = item.find('span', class_= 'article-title entry-title')
            item_b = item_a.find('a',class_='black-c')
            item_link = item_b.get('href')
            item_title = item_b.get('title')
#append in articles a list of dictionary 
            articles.append({'Aticle title':item_title,
                            'Article link': item_link})
for article in articles:
    print(article)
    
      




```
